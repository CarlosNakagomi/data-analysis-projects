UNIVARIATE ANALYSIS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Distribution of Numerical Features with Histogram, Boxplot, Violin Plot, and Normality Check
def univariate_analysis(df):
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
    for col in numeric_cols:
        if df[col].dropna().empty:
            print(f"Column {col} is empty or contains only NaN values. Skipping plot.")
            continue
        
        plt.figure(figsize=(12, 4))

        # Histogram & Density Plot
        plt.subplot(1, 4, 1)
        sns.histplot(df[col], kde=True)
        plt.title(f'Distribution of {col}')

        # Boxplot for Outlier Detection
        plt.subplot(1, 4, 2)
        sns.boxplot(x=df[col])
        plt.title(f'Boxplot of {col}')

        # Violin Plot for Outlier Detection
        plt.subplot(1, 4, 3)
        sns.violinplot(x=df[col])
        plt.title(f'Violin Plot of {col}')
        
        # Q-Q Plot for Normality Check
        plt.subplot(1, 4, 4)
        stats.probplot(df[col].dropna(), dist="norm", plot=plt)
        plt.title(f'Q-Q Plot of {col}')

        plt.tight_layout()
        plt.show()

        # Normality Test (Shapiro-Wilk Test)
        stat, p_value = stats.shapiro(df[col].dropna())
        if p_value > 0.05:
            print(f"Column {col} appears to follow a normal distribution (p-value: {p_value:.4f})")
        else:
            print(f"Column {col} does not appear to follow a normal distribution (p-value: {p_value:.4f})")

univariate_analysis(df)
```

```python
# Handling Outliers using IQR method
numeric_cols = df.select_dtypes(include='number').columns
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap outliers at lower and upper bounds
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
```

CHI-SQUARE

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Assuming the dataset is loaded as df
# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Chi-Square Test Function
def chi_square_test(df, target):
    significant_features = []
    
    for col in categorical_cols:
        contingency_table = pd.crosstab(df[col], df[target])
        chi2, p, dof, expected = chi2_contingency(contingency_table)

        # Display results
        print(f"\nFeature: {col}")
        print(f"Chi-Square Statistic: {chi2:.4f}")
        print(f"P-value: {p:.4f}")

        # Collect significant features (p < 0.05)
        if p < 0.05:
            significant_features.append(col)

    return significant_features

# Perform Chi-Square Test for each target variable
print("### Chi-Square Test for Acute Kidney Injury Outcome ###")
significant_features_aki = chi_square_test(df, 'Acute_Kidney_Injury_Outcome')

print("\n### Chi-Square Test for Dialysis Treatment ###")
significant_features_dialysis = chi_square_test(df, 'Dialysis_Treatment')

# Display significant features
print("\nSignificant Features for AKI Outcome:", significant_features_aki)
print("Significant Features for Dialysis Treatment:", significant_features_dialysis)
```

ANOVA

```python
from scipy.stats import f_oneway

# Identify numerical features
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# ANOVA Test Function
def anova_test(df, target):
    significant_features = []

    for col in numerical_cols:
        # Ensure the target is categorical
        groups = [df[col][df[target] == category] for category in df[target].unique()]

        # Perform ANOVA
        stat, p = f_oneway(*groups)

        print(f"\nFeature: {col}")
        print(f"ANOVA Statistic: {stat:.4f}")
        print(f"P-value: {p:.4f}")

        # Collect significant features
        if p < 0.05:
            significant_features.append(col)

    return significant_features

# Perform ANOVA Test for each target variable
print("### ANOVA Test for Acute Kidney Injury Outcome ###")
significant_features_aki = anova_test(df, 'Acute_Kidney_Injury_Outcome')

print("\n### ANOVA Test for Dialysis Treatment ###")
significant_features_dialysis = anova_test(df, 'Dialysis_Treatment')

# Display significant features
print("\nSignificant Features for AKI Outcome:", significant_features_aki)
print("Significant Features for Dialysis Treatment:", significant_features_dialysis)
```

MUTUAL INFORMATION

```python
from sklearn.feature_selection import mutual_info_classif
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Drop irrelevant column
df.drop('Unnamed: 25', axis=1, inplace=True)

# Encode categorical target variables
df['Acute_Kidney_Injury_Outcome'] = LabelEncoder().fit_transform(df['Acute_Kidney_Injury_Outcome'])
df['Dialysis_Treatment'] = LabelEncoder().fit_transform(df['Dialysis_Treatment'])

# Identify numerical features
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Mutual Information Function
def mutual_info_test(df, target):
    # Calculate MI scores
    mi_scores = mutual_info_classif(df[numerical_cols], df[target])

    # Display MI scores
    mi_df = pd.DataFrame({'Feature': numerical_cols, 'MI_Score': mi_scores})
    mi_df.sort_values(by='MI_Score', ascending=False, inplace=True)

    return mi_df

# Perform Mutual Information Test for each target variable
print("### Mutual Information for Acute Kidney Injury Outcome ###")
mi_scores_aki = mutual_info_test(df, 'Acute_Kidney_Injury_Outcome')
print(mi_scores_aki)

print("\n### Mutual Information for Dialysis Treatment ###")
mi_scores_dialysis = mutual_info_test(df, 'Dialysis_Treatment')
print(mi_scores_dialysis)
```

RELIEF

```python
from skrebate import ReliefF
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Encode categorical target variables
df['Acute_Kidney_Injury_Outcome'] = LabelEncoder().fit_transform(df['Acute_Kidney_Injury_Outcome'])
df['Dialysis_Treatment'] = LabelEncoder().fit_transform(df['Dialysis_Treatment'])

# Identify numerical features
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# ReliefF Feature Selection Function
def relief_test(df, target):
    # Initialize ReliefF model
    relief = ReliefF(n_features_to_select=len(numerical_cols))
    
    # Fit and rank features
    relief.fit(df[numerical_cols].values, df[target].values)
    
    # Display Feature Scores
    relief_scores = pd.DataFrame({
        'Feature': numerical_cols,
        'ReliefF_Score': relief.feature_importances_
    }).sort_values(by='ReliefF_Score', ascending=False)

    return relief_scores

# Perform ReliefF for each target variable
print("### ReliefF for Acute Kidney Injury Outcome ###")
relief_scores_aki = relief_test(df, 'Acute_Kidney_Injury_Outcome')
print(relief_scores_aki)

print("\n### ReliefF for Dialysis Treatment ###")
relief_scores_dialysis = relief_test(df, 'Dialysis_Treatment')
print(relief_scores_dialysis)
```

Justification for Feature Selection for Dialysis Treatment Prediction

The selected features for predicting Dialysis Treatment were chosen based on multiple statistical and machine learning feature selection techniques, ensuring robustness and strong predictive power. The following criteria were used to justify the selected features:

1. ANOVA Test (Analysis of Variance)

The ANOVA test identifies features that significantly differ between target classes. Features with p-values less than 0.05 were selected, as these indicate strong statistical significance.

Features like Thomay, Glasgow, Mach, HATB, SOFA, APACHEII, pH, HCO3, Lactac0, Urea_Level, Creatinine_Level, PCT0, BiLIrubin, Albumin_Level, and BC0 met this threshold, indicating strong correlation with the Dialysis Treatment outcome.

2. Mutual Information (MI)

MI measures the dependency between features and the target variable. Features with higher MI values are better predictors.

Features such as Thomay, HATB, Urea_Level, Creatinine_Level, and PCT0 displayed notable MI scores, reinforcing their importance in predicting Dialysis Treatment.

3. ReliefF Algorithm

ReliefF identifies relevant features by considering feature values in neighboring samples. It is effective in detecting feature interactions and handling noisy data.

Features like Thomay, Glasgow, HATB, SOFA, APACHEII, pH, and Creatinine_Level had high ReliefF scores, demonstrating their contribution to accurate model performance.

4. Random Forest Feature Importance

Random Forest models were used to assess each feature's contribution to improving model accuracy.

Features that consistently improved prediction accuracy were prioritized to ensure strong model performance.

5. Clinical Relevance

Some features like APACHEII, SOFA, and pH are widely recognized in medical literature as strong indicators of patient health status and organ function, making them essential for Dialysis Treatment prediction.

6. Feature Diversity

A mix of physiological (e.g., HATB, pH) and biochemical markers (e.g., Creatinine_Level, Urea_Level) was selected to ensure comprehensive coverage of factors influencing dialysis treatment.
